```python
import json
```

```python
db_info = json.load(open('./data/db_info.json'))
db_info
```

```python
# iterate over the keys & values in the dictionary

restructured_db_info = []


def compile_foreign_keys(db_info):
    fk = []
    for table_info in db_info:
        table_name = table_info["table_name"]
        foreign_keys = table_info["foreign_keys"]
        for foreign_key in foreign_keys:
            foreign_key["referenced"] = table_name
            fk.append(foreign_key)
    return fk.copy()


for table_info_o in db_info:
    table_info = table_info_o.copy()
    table_name = table_info["table_name"]

    # Get all the times the table was in a foreign key
    foreign_keys = [
        fk for fk in compile_foreign_keys(db_info) if fk["table"] == table_name
    ]

    table_info["foreign_keys"] = []

    # Delete the 'table' key from the foreign key
    for foreign_key_0 in foreign_keys:
        # Rename the 'referenced' key to 'table'
        foreign_key = foreign_key_0.copy()
        foreign_key["table"] = foreign_key["referenced"]
        del foreign_key["referenced"]
        table_info["foreign_keys"].append(foreign_key)
    # Add the foreign keys to the table_info

    # Add the table_info to the restructured_db_info
    restructured_db_info.append(table_info)

restructured_db_info
```

```python
# Bring to db_info.json
with open("./data/restructured_db_info.json", "w") as outfile:
    json.dump(restructured_db_info, outfile, indent=4)
```
